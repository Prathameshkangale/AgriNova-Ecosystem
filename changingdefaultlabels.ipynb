```python
import os

# ── CONFIG ─────────────────────────────────────────────────
DATASET_PATH = r"C:\Users\anilk\OneDrive\Desktop\mit3"
SUBFOLDERS   = ["train", "valid"]          # folders to process
OLD_CLASS    = "0"                         # class index to replace
NEW_CLASS    = "4"                         # new class index
# ───────────────────────────────────────────────────────────

def fix_label_file(filepath):
    """Read a YOLO label file and replace first-column 0 → 4."""
    with open(filepath, "r") as f:
        lines = f.readlines()

    new_lines = []
    changed   = False

    for line in lines:
        stripped = line.strip()
        if not stripped:                   # skip blank lines
            new_lines.append(line)
            continue

        parts = stripped.split()
        if parts[0] == OLD_CLASS:          # only touch class-0 entries
            parts[0] = NEW_CLASS
            new_lines.append(" ".join(parts) + "\n")
            changed = True
        else:
            new_lines.append(line)

    if changed:
        with open(filepath, "w") as f:
            f.writelines(new_lines)

    return changed


def process_folder(folder_path):
    """Walk a labels/ subfolder and fix every .txt file."""
    labels_dir = os.path.join(folder_path, "labels")

    if not os.path.isdir(labels_dir):
        print(f"  ⚠️  No 'labels' subfolder found in: {folder_path}")
        return 0, 0

    total  = 0
    fixed  = 0

    for root, _, files in os.walk(labels_dir):
        for fname in files:
            if not fname.endswith(".txt"):
                continue
            fpath  = os.path.join(root, fname)
            total += 1
            if fix_label_file(fpath):
                fixed += 1
                print(f"  ✅ Fixed : {fpath}")

    return total, fixed


def main():
    print("=" * 60)
    print("  YOLO Label Class-Index Fixer  (0 → 4)")
    print("=" * 60)
    print(f"  Dataset : {DATASET_PATH}\n")

    grand_total = 0
    grand_fixed = 0

    for sub in SUBFOLDERS:
        folder = os.path.join(DATASET_PATH, sub)
        if not os.path.isdir(folder):
            print(f"⚠️  Skipping '{sub}' — folder not found: {folder}")
            continue

        print(f"\n📂 Processing: {folder}")
        total, fixed = process_folder(folder)
        grand_total += total
        grand_fixed += fixed
        print(f"   → {fixed}/{total} label files updated in '{sub}'")

    print("\n" + "=" * 60)
    print(f"  DONE!  {grand_fixed}/{grand_total} files updated in total.")
    print("=" * 60)


if __name__ == "__main__":
    main()
```
