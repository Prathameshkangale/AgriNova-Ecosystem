```python
import zipfile, os, shutil

print("Starting extraction process...")

# Ensure the target directory is clean before extraction
target_dir = "/content/ag/"
if os.path.exists(target_dir):
    print(f"Removing existing directory: {target_dir}")
    shutil.rmtree(target_dir)
    print(f"Directory {target_dir} removed.")
else:
    print(f"Directory {target_dir} does not exist. No need to remove.")

print(f"Creating target directory: {target_dir}")
os.makedirs(target_dir, exist_ok=True)
print(f"Directory {target_dir} created.")

zip_file_path = "ag.zip"
if not os.path.exists(zip_file_path):
    print(f"Error: Zip file {zip_file_path} not found!")
else:
    print(f"Extracting {zip_file_path} to {target_dir}...")
    try:
        with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
            zip_ref.extractall(target_dir)
        print("✅ Extraction complete!")
    except Exception as e:
        print(f"Error during extraction: {e}")

print(f"Contents of {target_dir}: {os.listdir(target_dir)}")

# Verify contents of the nested 'ag' directory
nested_ag_dir = os.path.join(target_dir, "ag")
if os.path.exists(nested_ag_dir):
    print(f"Contents of {nested_ag_dir} after re-extraction: {os.listdir(nested_ag_dir)}")
else:
    print(f"{nested_ag_dir} does not exist after re-extraction. Something is still wrong.")
```

```python
import shutil, os

# Correcting the paths based on the confirmed nested structure: /content/ag/ag/...
shutil.move("/content/ag/ag/train", "/content/ag/train")
shutil.move("/content/ag/ag/valid", "/content/ag/valid")
shutil.move("/content/ag/ag/data12.yaml", "/content/ag/data12.yaml")

print("✅ Fixed!")
print(os.listdir("/content/ag/"))
```

```python
from ultralytics import YOLO

model = YOLO("yolov8s.pt")

model.train(
    data="/content/ag/data12.yaml",
    epochs=100,
    imgsz=640,
    batch=16,
    name="ag_pest_v1",
    project="/content/runs",
    patience=20,
    augment=True,
    hsv_h=0.015,
    hsv_s=0.7,
    flipud=0.3,
    fliplr=0.5,
    save_period=10
)
```

```python
metrics = model.val()
print(f"mAP50:     {metrics.box.map50:.3f}")
print(f"mAP50-95:  {metrics.box.map:.3f}")
print(f"Precision: {metrics.box.mp:.3f}")
print(f"Recall:    {metrics.box.mr:.3f}")
```
